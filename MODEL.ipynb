```python
import ast

def convert2(obj):
    L = []
    try:
        # Check if it's a string before evaluating
        if isinstance(obj, str):
            data = ast.literal_eval(obj)
            for i in data:
                L.append(i['name'])
        return L
    except Exception as e:
        return []

# Apply the fix
merged['genres'] = merged['genres'].apply(convert2)

# Check the first few rows - they should no longer be empty!
print(merged['genres'].head())
```

```python
merged.head()
```

```python
merged['keywords'] = merged['keywords'].apply(convert2)
print(merged['keywords'].head())
```

```python
def convert3(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
merged['cast'] = merged['cast'].apply(convert3)
print(merged['cast'].head())
```

```python
merged.head()
```

```python
def fetch_director(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director' and counter!=1:
            L.append(i['name'])
            counter+=1
            break
    return L
merged['crew'] = merged['crew'].apply(fetch_director)
print(merged['crew'].head())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df=merged[['id','title','tags']]
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

# Vectorization

```python

from sklearn.feature_extraction.text import CountVectorizer
model = CountVectorizer(max_features=5000, stop_words='english')
vectors = model.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
model.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
distance=cosine_similarity(vectors)
```

```python
distance
```

```python
sorted(list(enumerate(distance[0])), reverse=True, key=lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = distance[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
recommend('Batman Begins')
```
